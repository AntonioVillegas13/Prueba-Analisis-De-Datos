```python
#LIBRERIAS A IMPORTAR
from bs4 import BeautifulSoup
import requests
import couchdb
import json
from bson import json_util, ObjectId


```

```python
urlBase = "https://puregaming.es/noticias-videojuegos/"
maxPages = 20
counter = 0
quotes = []
for i in range(1,maxPages):

    # Construyo la URL
    if i > 1:
        url = "%spage/%d/" %(urlBase,i)
    else:
        url = urlBase

    # Realizamos la petición a la web
    req = requests.get(url)
    # Comprobamos que la petición nos devuelve un Status Code = 200
    statusCode = req.status_code
    if statusCode == 200:

        # Pasamos el contenido HTML de la web a un objeto BeautifulSoup()
        html = BeautifulSoup(req.text)

        # Obtenemos todos los divs donde estan las entradas
        entradas = html.find_all('div',{'class':'post post-styles post-style-3'})

        # Recorremos todas las entradas para extraer el título, autor y fecha
        for entrada in entradas:
            quote = {}
            counter += 1
            titulo = entrada.find('div', {'class' : 'title'}).getText()
            contenido = entrada.find('div', {'class' : 'summary'}).getText()
            autor = entrada.find('li', {'class' : 'author'}).getText()
            categoria = entrada.find('li', {'class' : 'category'}).getText()
            fecha = entrada.find('li', {'class' : 'date'}).getText()
            quote['titulo'] = titulo
            quote['contenido'] = contenido
            quote['author'] = autor
            quote['categoria'] = categoria
            quote['fecha'] = fecha
            
            #Imprimo el Título, Autor y Fecha de las entradas
            print ("\n%d - %s -%s - %s -%s" %(counter,titulo,autor,categoria,fecha))
            print("\n%s" % (contenido)) 
            quotes.append(quote) 
            
    else:
        # Si ya no existe la página y me da un 400
        break

        

```
