```python
#LIBRERIAS A IMPORTAR
from bs4 import BeautifulSoup
import requests
from pymongo import MongoClient
```

```python
#CONEXION A MONGO DB
try:
        client = MongoClient('mongodb://localhost:27017')
        client.server_info()
        print("Conexion exitosa")
        client.close
except pymongo.errors.ServerSelectionTimeoutError as errorTiempo:
        print("Conexi√≥n rechazada")
        
#CREACION DE BASE Y COLECCION        
DB = client['PUREGAME']
col = DB['articulos']
```

```python
#WEB SCRAPING CON BEAUTIFULSOUP
urlBase = "https://puregaming.es/noticias-videojuegos/"
maxPages = 20
counter = 0
quotes = []
for i in range(1,maxPages):

    # Construyo la URL
    if i > 1:
        url = "%spage/%d/" %(urlBase,i)
    else:
        url = urlBase

    # Realizamos la petici√≥n a la web
    req = requests.get(url)
    # Comprobamos que la petici√≥n nos devuelve un Status Code = 200
    statusCode = req.status_code
    if statusCode == 200:

        # Pasamos el contenido HTML de la web a un objeto BeautifulSoup()
        html = BeautifulSoup(req.text)

        # Obtenemos todos los divs donde estan las entradas
        entradas = html.find_all('div',{'class':'post post-styles post-style-3'})

        # Recorremos todas las entradas para extraer el t√≠tulo, autor y fecha
        for entrada in entradas:
            quote = {}
            counter += 1
            titulo = entrada.find('div', {'class' : 'title'}).getText()
            contenido = entrada.find('div', {'class' : 'summary'}).getText()
            autor = entrada.find('li', {'class' : 'author'}).getText()
            categoria = entrada.find('li', {'class' : 'category'}).getText()
            fecha = entrada.find('li', {'class' : 'date'}).getText()
            quote['titulo'] = titulo
            quote['contenido'] = contenido
            quote['author'] = autor
            quote['categoria'] = categoria
            quote['fecha'] = fecha
            
            #Imprimo el T√≠tulo, Autor y Fecha de las entradas
            print ("\n%d - %s -%s - %s -%s" %(counter,titulo,autor,categoria,fecha))
            print("\n%s" % (contenido)) 
            quotes.append(quote) 
            
    else:
        # Si ya no existe la p√°gina y me da un 400
        break
try:
    col.insert_many(quotes)
    print("Informaci√≥n grabada en la base de datos correctamente" )
    
    
except:
    print('Se ha producido un error los datos no fueron grabados en la base de datos')     
```
