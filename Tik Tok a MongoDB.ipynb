```python
#el archivo json debe estar en la misma carpeta que el proyecto
df1 = pd.read_json("videojuegos.json",encoding="utf8")    
df1.head()
```

```python
try:
    client = MongoClient('mongodb://localhost:27017')
    client.server_info()
    print("Conexion exitosa")
    client.close
    
except pymongo.errors.ServerSelectionTimeoutError as errorTiempo:
    print("Conexi√≥n rechazada")
dbm = client['tiktok']
cole = dbm['tiktoktest']

```

```python
def createDocsFromDF(df1, collection = None, insertToDB=False):
    docs = []
    fields = [col for col in df1.columns]
    for i in range(len(df1)):
        doc = {col:df1[col][i] for col in df1.columns if col != 'index'}
        for key, val in doc.items():
            if type(val) == np.int64:
                doc[key] = int(val)
            if type(val) == np.float64:
                doc[key] = float(val)
            if type(val) == np.bool_:
                doc[key] = bool(val)
        docs.append(doc) 
    if isinstance(docs, list): 
        cole.insert_many(docs)   
    else: 
        cole.insert_one(docs)
    print("guardado exitosamente")    
    return docs 
```
